## Data Preparation

```python
# Replace outliers with interpolation
def identify_outliers(df):
    outliers_dict = {}

    for (sku, supermarket), group in df.groupby(['sku', 'supermarket']):
        # Calculate IQR (Interquartile Range)
        Q1 = group['demand'].quantile(0.25)
        Q3 = group['demand'].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 3 * IQR
        upper_bound = Q3 + 3 * IQR
        
        outliers = group[(group['demand'] < lower_bound) | (group['demand'] > upper_bound)]
        outliers_dict[(sku, supermarket)] = outliers.index.tolist()

    return outliers_dict
    
def replace_outliers_with_interpolation(df):
    # Find outliers
    outliers_dict = identify_outliers(df)

    # Replace outliers with interpolaction
    for indices in outliers_dict.values():
        df.loc[indices, 'demand'] = np.nan

    df['demand'] = df['demand'].interpolate(method='linear')  # Linear interpolaton

    return df

df_interp = df.copy()
df_interp = replace_outliers_with_interpolation(df_interp)
```

```python
# Aggregete data weekly
grouped = df_interp.groupby(["sku", "supermarket"])
# Performs a simplistic aggregation of promotion. 
# If a promotion occured during the week this variable will return the number of days the promotion occurs.
df_weekly = grouped.apply(lambda df: df.resample("W").agg(demand=('demand', 'sum'),
                                                          promotion=('promotion', 'sum'),
                                                          promotion_flag=('promotion', 'max')))
df_weekly = df_weekly.reset_index().set_index("date")
```

```python
# Plot the result of the aggregation and outliers replacement 
def plot_demand_with_promotion(group, sku, supermarket, ax):
    group = group.reset_index()

    # Draw the entire line in blue (without distinction on the page)
    sns.lineplot(
        x='date', 
        y='demand', 
        data=group, 
        color='blue', 
        label='Demand', 
        ax=ax
    )
    
    # Highlight in red the points where `promotion == True`
    sns.scatterplot(
        x='date', 
        y='demand', 
        data=group[group['promotion'] != 0], 
        color='red', 
        label='Promotion', 
        ax=ax
    )

    ax.set_title(f'{sku} in {supermarket}')
    ax.set_xlabel('Date')
    ax.set_ylabel('Demand')

grouped = df_weekly.groupby(['sku', 'supermarket'])

fig, axes = plt.subplots(9, 1, figsize=(16, 40))
axes = axes.flatten()

for ax, ((sku, supermarket), group) in zip(axes, grouped):
    plot_demand_with_promotion(group, sku, supermarket, ax)

plt.tight_layout()
plt.show()
```

In the daily data analysis, we could clearly observe trends and seasonality. To avoid losing important daily-level insights, it might be useful to introduce a new feature that captures daily variability, such as the maximum change in daily demand within a week.

Additionally, the demand values for the boundary weeks (beginning and end) are underestimated because they don't represent a full week of data. A more accurate approach would involve multiplying these values by a scalar to approximate a full week (e.g., if the first week has only 3 days, we would multiply the demand by 7/3). However, for the purposes of this analysis, I will simplify the approach by removing these partial weeks from the dataset.

```python
# Delete rows with minimum and maximum index
def remove_min_max_index(group):
    min_index = group.index.min()
    max_index = group.index.max()
    return group.drop([min_index, max_index])

df_weekly = df_weekly.groupby(['sku', 'supermarket'], group_keys=False).apply(remove_min_max_index)
```

```python
# Plot the result of the aggregation and outliers replacement 
def plot_demand_with_promotion(group, sku, supermarket, ax):
    group = group.reset_index()

    # Draw the entire line in blue (without distinction on the page)
    sns.lineplot(
        x='date', 
        y='demand', 
        data=group, 
        color='blue', 
        label='Demand', 
        ax=ax
    )
    
    # Highlight in red the points where `promotion == True`
    sns.scatterplot(
        x='date', 
        y='demand', 
        data=group[group['promotion'] != 0], 
        color='red', 
        label='Promotion', 
        ax=ax
    )

    ax.set_title(f'{sku} in {supermarket}')
    ax.set_xlabel('Date')
    ax.set_ylabel('Demand')

grouped = df_weekly.groupby(['sku', 'supermarket'])

fig, axes = plt.subplots(9, 1, figsize=(16, 40))
axes = axes.flatten()

for ax, ((sku, supermarket), group) in zip(axes, grouped):
    plot_demand_with_promotion(group, sku, supermarket, ax)

plt.tight_layout()
plt.show()
```

I will now proceed to add new date-related features, including year, month, week of the year, and quarter. Following that, I will introduce lag variables, specifically lag_1, lag_2, lag_4, and lag_8, to capture historical demand patterns. Additionally, I will shift the target variable by 8 weeks forward to align with our prediction goal. The same process will be applied to the promotion variable.

Moreover, it would be beneficial to consider incorporating additional features, such as a moving average. The moving average can help smooth out short-term fluctuations and highlight longer-term trends in demand. Other potential features to explore could include:

* Exponential Moving Average (EMA): This places greater weight on more recent observations, which can be more responsive to changes in demand patterns.
* Cyclical Features: Variables such as the number of days since the last promotion or previous peaks in demand can provide insights into cyclical purchasing behavior.
  
By enriching the dataset with these features, we aim to enhance the models predictive power and better understand the factors influencing demand.

```python
# Add new date features
df_weekly['year'] = df_weekly.index.year
df_weekly['month'] = df_weekly.index.month
df_weekly['week_of_year'] = df_weekly.index.isocalendar().week
df_weekly['quarter'] = df_weekly.index.quarter
```

```python
# Add lag features
def prepare_forecast_data(group, weeks_ahead=8):
    
    group['lag_1'] = group['demand'].shift(1) 
    group['lag_2'] = group['demand'].shift(2)
    group['lag_4'] = group['demand'].shift(4)
    group['lag_8'] = group['demand'].shift(8)
    group['promotion_target'] = group['promotion'].shift(-weeks_ahead)
    group['promotion_flag_target'] = group['promotion_flag'].shift(-weeks_ahead).astype(bool)
    
    # Fill in NaN in lags with demand value
    group.fillna(method='bfill', inplace=True)

    group['target_demand'] = group['demand'].shift(-weeks_ahead)
    group.rename(columns={'demand': 'this_week_demand',
                         'promotion': 'this_week_promotion'}, inplace=True)
    group.dropna(subset=['target_demand'], inplace=True)

    return group
    
results = []

for (sku, supermarket), group in df_weekly.groupby(['sku', 'supermarket']):
    group_with_lags = prepare_forecast_data(group.copy())
    results.append(group_with_lags)

df_target = pd.concat(results)
df_target.drop(columns=['this_week_promotion', 'promotion_flag'], inplace=True)
```

### Train Test Split

```python
# Select split date
unique_index_list = df_target.index.unique().sort_values()
cutoff_index = np.ceil(len(unique_index_list) * 0.8)
cutoff_date = unique_index_list[int(cutoff_index)]
```

```python
# Use One-Hot Encoding for 'sku' and 'supermarket' columns
df_target = pd.get_dummies(df_target, columns=['sku', 'supermarket'])
```

```python
train_data = df_target[df_target.index <= cutoff_date]
test_data = df_target[df_target.index > cutoff_date]

X_train = train_data.drop(columns=['target_demand'])
y_train = train_data['target_demand']

X_test = test_data.drop(columns=['target_demand'])
y_test = test_data['target_demand']

X_train.head(3)
```

### XGBoost Model

```python
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42, eval_metric="mae")
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
}

grid_search = GridSearchCV(estimator=xgb_model, 
                           param_grid=param_grid, 
                           scoring='neg_mean_squared_error',
                           n_jobs=-1)
grid_search.fit(X_train, 
                y_train,
                eval_set=[(X_test, y_test)], 
                verbose=False)
```

```python
print("Best parameters found: ", grid_search.best_params_)
```

### Validation

```python
tscv = TimeSeriesSplit(n_splits=5)

# Validation
for train_index, val_index in tscv.split(X_train):
    X_train_cv, X_val_cv = X_train.iloc[train_index], X_train.iloc[val_index]
    y_train_cv, y_val_cv = y_train.iloc[train_index], y_train.iloc[val_index]
    
    grid_search.fit(X_train_cv, y_train_cv)
    
    y_val_pred = grid_search.predict(X_val_cv)
    
    mae = mean_absolute_error(y_val_cv, y_val_pred)
    print(f"Mean Absolute Error: {round(mae, 2)}.")
    
print("Best parameters found: ", grid_search.best_params_)
```

```python
# Take the best model
best_model = grid_search.best_estimator_

sku_columns = ['sku_desperados', 'sku_heineken 0.0', 'sku_heineken regular']
supermarket_columns = ['supermarket_albert-heijn', 'supermarket_dirk', 'supermarket_jumbo']

features = ['this_week_demand', 'year', 'month', 'week_of_year', 'quarter', 
            'lag_1', 'lag_2', 'lag_4', 'lag_8', 'promotion_target', 'promotion_flag_target'] + sku_columns + supermarket_columns

def calculate_mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def plot_group(test_data, df, group_name):
    plt.figure(figsize=(10, 6))
    
    # Plot the real demand
    plt.plot(df.index, df['target_demand'], label='Actual Demand', color='blue')

    # Plot predicted demand
    plt.plot(test_data.index, test_data['predicted_demand'], label='Predicted Demand', color='orange', linestyle='--')

    # Heighlight promotion in red
    promotions = df[df['promotion_target'] > 0]
    plt.scatter(promotions.index, promotions['target_demand'], color='red', label='Promotion', s=60)

    # Add cutoff line
    plt.axvline(x=cutoff_date, color='black', linestyle='--', label='Cutoff Date')

    plt.title(f'Demand Prediction for {group_name}')
    plt.xlabel('Date')
    plt.ylabel('Demand')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()

def train_and_predict_group(df, group_name):
    # Prepare X and y for testing
    test_data = df[df.index > cutoff_date]
    X = test_data[features]
    y = test_data['target_demand']
    
    # Prediction
    test_data['predicted_demand'] = best_model.predict(X)
    mape = calculate_mean_absolute_percentage_error(test_data['target_demand'], 
                                                    test_data['predicted_demand'])
    print(f"\n Mean Absolute Percentage Error for {group_name}: {round(mape, 2)} % \n")

    plot_group(test_data, df, group_name)

# Loop over each SKU & supermarket pair
for sku_col in sku_columns:
    for supermarket_col in supermarket_columns:
        group_data = df_target[(df_target[sku_col] == True) & (df_target[supermarket_col] == True)]
                
        if not group_data.empty:
            group_name = f"{sku_col} - {supermarket_col}"
            train_and_predict_group(group_data, group_name)
```

The Mean Absolute Error (MAE) metric provides an intuitive way for business stakeholders to understand how well the model performs. It calculates the average quantity of errors in the predictions, without considering their direction (whether the prediction is higher or lower than the actual value). This makes it a useful measure for demand forecasting because it directly translates into units of demand  in other words, how many units off our forecast might be, on average, from the actual demand.

From a business perspective, this has clear implications:

1) Lower MAE means more accurate predictions, allowing better inventory management, preventing overstock or stockouts, and leading to reduced costs.
2) If MAE is high, it indicates the model might not be accurately predicting future demand, leading to operational inefficiencies like excess inventory (which ties up capital and leads to markdowns) or stockouts (which result in missed sales opportunities and lower customer satisfaction).

### Feature Importance Calculation and Visualization

```python
# Extract feature importance from the trained model
importance = best_model.get_booster().get_score(importance_type='weight')
importance_df = pd.DataFrame({
    'Feature': importance.keys(),
    'Importance': importance.values()
})
# Sort by importance to show the most important features first
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 6))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.xlabel('Feature Importance Score')
plt.title('Feature Importance in Predicting Demand (8 Weeks in Advance)')
plt.gca().invert_yaxis()  # Invert y-axis to have the most important feature at the top
plt.show()
```

1) This Week's Demand: The most important feature in predicting future demand is the immediate past demand, which shows that current trends are highly indicative of near-future patterns.
2) Week of Year: Seasonality plays a crucial role, as certain times of the year (like holidays or summer months) consistently see higher or lower demand.
3) Lag Features: The demand from one, two, four, and eight weeks ago are important as they provide historical trends, showing how recurring weekly demand patterns affect future forecasts.
4) Promotions: The presence or absence of a promotion has impact on demand, as it can lead to spikes.
5) SKU and Supermarket: Different SKUs and supermarkets have distinct demand behaviors, which means adjusting forecasts based on these variables is essential.
