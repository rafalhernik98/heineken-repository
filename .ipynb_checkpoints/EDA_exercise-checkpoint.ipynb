#### Promotion Impact and Outlier Identification

```python
# Function to create a chart highlighting promotions
def plot_demand_with_promotion(group, sku, supermarket, ax):
    group = group.reset_index()

    # Draw the entire line in blue (without distinction on the page)
    sns.lineplot(
        x='date', 
        y='demand', 
        data=group, 
        color='blue', 
        label='Demand', 
        ax=ax
    )
    
    # Highlight in red the points where `promotion == True`
    sns.scatterplot(
        x='date', 
        y='demand', 
        data=group[group['promotion'] == True], 
        color='red', 
        label='Promotion', 
        ax=ax
    )


    ax.set_title(f'{sku} in {supermarket}')
    ax.set_xlabel('Date')
    ax.set_ylabel('Demand')

grouped = df.groupby(['sku', 'supermarket'])

fig, axes = plt.subplots(9, 1, figsize=(16, 40))
axes = axes.flatten()

for ax, ((sku, supermarket), group) in zip(axes, grouped):
    plot_demand_with_promotion(group, sku, supermarket, ax)

plt.tight_layout()
plt.show()
```

There is a clear increase in demand on promotion days. However, it is not very visible due to outliers that change the scale of the graphs.

There are many outliers in each group that can be captured with the code below.

```python
def identify_outliers(df):
    outliers_dict = {}

    for (sku, supermarket), group in df.groupby(['sku', 'supermarket']):
        # Calculate IQR (Interquartile Range)
        Q1 = group['demand'].quantile(0.25)
        Q3 = group['demand'].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 3 * IQR
        upper_bound = Q3 + 3 * IQR
        
        outliers = group[(group['demand'] < lower_bound) | (group['demand'] > upper_bound)]
        outliers_dict[(sku, supermarket)] = outliers.index.tolist()

    return outliers_dict

outliers = identify_outliers(df)
outlier_dates = []
for key, value in outliers.items():
    print(f"Outliers for {key}:")
    print(value)
    # outlier_dates_in_group = np.datetime_as_string(value.index.values, unit='D')
    outlier_dates_in_group = value
    outlier_dates.extend(outlier_dates_in_group)
    
date_counts = {}

for date in outlier_dates:
    if date in date_counts:
        date_counts[date] += 1
    else:
        date_counts[date] = 1

sorted_date_counts = sorted(date_counts.items(), key=lambda item: (-item[1], item[0]))

for date, count in sorted_date_counts:
    print(f"{date}: {count}")
```

After getting rid of the outliers, we can see a trend in the data and a clear increase in demand during the promotion.

However, removing outliers completely is not the best approach, as we now have gaps in the date index. Instead of removing them completely, we will replace them with interpolation.

```python
def replace_outliers_with_interpolation(df):
    # Find outliers
    outliers_dict = identify_outliers(df)

    # Replace outliers with interpolaction
    for indices in outliers_dict.values():
        df.loc[indices, 'demand'] = np.nan

    df['demand'] = df['demand'].interpolate(method='linear')  # Linear interpolaton

    return df, outliers_dict
```

```python
def plot_demand_with_outliers(group, sku, supermarket, outlier_indices, ax):
    group = group.reset_index()

    sns.lineplot(
        x='date', 
        y='demand', 
        data=group, 
        color='blue', 
        label='Interpolated Demand', 
        ax=ax
    )
    sns.scatterplot(
        x='date', 
        y='demand', 
        data=group[group['promotion'] == True], 
        color='red', 
        label='Promotion', 
        ax=ax,
        s=80  
    )
           
    outliers = group[group['date'].isin(outlier_indices)]
    sns.scatterplot(
        x='date', 
        y='demand', 
        data=outliers, 
        color='black', 
        label='Outliers (Replaced)', 
        ax=ax,
        s=80
    )

    ax.set_title(f'{sku} in {supermarket}')
    ax.set_xlabel('Date')
    ax.set_ylabel('Demand')
```

```python
df_interp = df.copy()
df_interp, outliers_dict = replace_outliers_with_interpolation(df_interp)

grouped = df_interp.groupby(['sku', 'supermarket'])
fig, axes = plt.subplots(len(grouped), 1, figsize=(12, len(grouped) * 4))  

for ax, ((sku, supermarket), group) in zip(axes, grouped):
    outlier_indices = pd.Index(outliers_dict[(sku, supermarket)]) 
    plot_demand_with_outliers(group, sku, supermarket, outlier_indices, ax)

plt.tight_layout()
plt.show()
```

Some conclusions can already be drawn from the graphs, but they will be more readable after smoothing the time series.

```python
# Smoothing
df_uni = df_interp.groupby('date')['demand'].sum()

model = SimpleExpSmoothing(df_uni.values)
model_fit = model.fit(smoothing_level=0.2, optimized=False) 

# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(df_uni.values, label='Actual Demand', color='blue')
plt.plot(model_fit.fittedvalues, label='Fitted Values', color='orange')
plt.title('Simple Exponential Smoothing')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.legend()
plt.show()
```

```python
# Initialize a dictionary to store fitted values
fitted_values = {}

# Group by 'sku' and 'supermarket' columns
for (sku, supermarket), group_data in df_interp.groupby(['sku', 'supermarket']):
    # Fit the Simple Exponential Smoothing model
    model = SimpleExpSmoothing(group_data['demand'])
    model_fit = model.fit(smoothing_level=0.2, optimized=False)  # Adjust smoothing level as necessary

    # Store fitted values
    fitted_values[(sku, supermarket)] = model_fit.fittedvalues

    # Plotting results for each group
    plt.figure(figsize=(12, 6))
    plt.plot(model_fit.fittedvalues, label='Fitted Values', color='orange', linewidth=2)
    
    # Highlight promotion days
    sns.scatterplot(
        x=group_data.index[group_data['promotion']],  # Get dates where promotion is True
        y=group_data['demand'][group_data['promotion']],
        color='green', 
        label='Promotion', 
        s=80  # Size of points
    )
    plt.title(f'Simple Exponential Smoothing Fitted Values for {sku} in {supermarket}')
    plt.xlabel('Date')
    plt.ylabel('Demand')
    plt.legend()
    plt.grid()
    plt.show()
```

1. Statistical Testing for Seasonality and Trends:

- Conduct specific statistical tests to assess the presence of seasonality and trends in the data. This will include:
Seasonal Decomposition: Using methods like STL (Seasonal and Trend decomposition using Loess) to visualize and quantify seasonal effects.
Dickey-Fuller Test: To check the stationarity of the time series. If the series is found to be non-stationary, appropriate transformations such as differencing will be applied to stabilize the mean and variance.
Analysis of Outliers:

2. Outliers and anomalous observations 
- Identifying outliers using statistical methods (e.g., Z-scores, IQR method) and examining their context.
- Attempting to correlate outliers with specific events (e.g., holidays, major local events, or promotional activities). However, preliminary analyses have not yielded clear associations, indicating that further investigation may be required to establish any potential links.

3. Model Selection Considerations:

An important part of this analysis will be deciding on the most appropriate modeling approach. Three potential strategies will be evaluated:

a) Single Large Model for All 9 Groups
- Advantages:

Simplicity: Easier to implement and maintain one comprehensive model rather than multiple.

Resource Efficiency: Reduces computational costs and time associated with model training and updates.

Unified Insights: Provides a holistic view of demand patterns across all groups, potentially capturing overall seasonal trends and relationships.

- Disadvantages:

Loss of Specificity: A single model may not adequately capture unique seasonal patterns or behaviors specific to individual groups, especially if demand patterns vary significantly.

Complexity in Interpretability: Understanding results and making targeted decisions may be more challenging due to aggregated data.

b) Separate Models for Each Group (9 Models)
- Advantages:

Tailored Insights: Each model can be optimized for the specific demand patterns and behaviors of individual groups, leading to more accurate forecasts.

Flexibility: Ability to adapt each model as the market changes or as specific groups exhibit different trends.

- Disadvantages:

Resource Intensive: Increased computational and time costs for developing, maintaining, and retraining 9 separate models.

Scalability Issues: As the number of stores or SKUs grows, managing multiple models could become unwieldy and costly.

c) Intermediate Approach with 3 Models for Different SKUs
- Advantages:

Balance: Strikes a balance between specificity and resource efficiency, allowing for tailored models while reducing overall complexity.

Focused Optimization: Different models can cater to varying demand characteristics among the most popular SKUs, which are likely to have more significant impacts on overall sales.

- Disadvantages:

Potential Overlap: There may still be missed opportunities for insights that could be derived from a broader, all-encompassing model.

Increased Management: While fewer than 9 models, managing multiple models can still require substantial effort and resources.

```python
# Prepare figure layout to have each supermarket with its own subplot
supermarkets = df_interp['supermarket'].unique()
skus = df_interp['sku'].unique()

# Create subplots for each supermarket
fig, axes = plt.subplots(nrows=len(supermarkets), ncols=1, figsize=(12, 6*len(supermarkets)), sharex=True)

# Delete rows with minimum and maximum index
def remove_min_max_index(group):
    min_index = group.index.min()
    max_index = group.index.max()
    return group.drop([min_index, max_index])


# Loop through each supermarket and create a subplot for the demand trends
for ax, supermarket in zip(axes, supermarkets):
    # Filter the data for the current supermarket
    df_supermarket = df_interp[df_interp['supermarket'] == supermarket]
    
    # Plot each SKU for the current supermarket on the same plot
    for sku in skus:
        df_sku = df_supermarket[df_supermarket['sku'] == sku]
        df_sku = df_sku.resample('W').sum()  # Resample demand data weekly
        df_sku = df_sku.groupby(['sku', 'supermarket'], group_keys=False).apply(remove_min_max_index)
        
        ax.plot(df_sku.index, df_sku['demand'], label=sku)

    # Set titles and labels
    ax.set_title(f'Weekly Demand Trends for {supermarket}', fontsize=14)
    ax.set_ylabel('Demand')
    ax.legend(title='SKU')
    ax.grid(True)

# Set common X label
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```
